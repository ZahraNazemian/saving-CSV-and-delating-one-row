```python
import pandas as pd
import requests


# Step 1: Download and save the CSV from the URL
csv_url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-white.csv'
response = requests.get(csv_url)

# Save the CSV file locally (name it 'winequality-white.csv')
with open('winequality-white.csv', 'wb') as file:
    file.write(response.content)

print("CSV file downloaded and saved as 'winequality-white.csv'")

# Step 2: Load the CSV file from local disk (without original headers)
df = pd.read_csv('winequality-white.csv', sep=';', skiprows=1)  # `sep=';'` since the columns are separated by semicolons

# Step 3: Define and assign new headers (custom headers as per your need)
new_headers = ['FixedAcidity', 'VolatileAcidity', 'CitricAcid', 'ResidualSugar', 'Chlorides',
               'FreeSulfurDioxide', 'TotalSulfurDioxide', 'Density', 'pH', 'Sulphates', 
               'Alcohol', 'Quality']  # These headers match the wine dataset's structure

# Assign the new headers
df.columns = new_headers

# Step 4: Display the DataFrame with new headers
df.head()  # Shows the first few rows of the DataFrame
```
